```python
import pandas as pd

# Load the CSV files with appropriate encoding and delimiter
player_stats_2021_2022 = pd.read_csv('raw_data/2021-2022 Football Player Stats.csv', encoding='latin1', delimiter=';')
player_stats_2022_2023 = pd.read_csv('raw_data/2022-2023 Football Player Stats.csv', encoding='latin1', delimiter=';')
player_valuations = pd.read_csv('raw_data/player_valuations.csv', encoding='latin1')
players = pd.read_csv('raw_data/players.csv', encoding='latin1')

player_stats_2021_2022.head(), player_stats_2022_2023.head(), player_valuations.head(), players.head()

```

```python

# Merge player valuations with players to get player names
player_valuations_with_names = player_valuations.merge(players[['player_id', 'name']], on='player_id', how='left')

# Extract relevant columns: player names and market valuations
player_valuations_relevant = player_valuations_with_names[['name', 'date', 'market_value_in_eur']]
player_valuations_relevant.columns = ['Player','date', 'Valuation']
player_valuations_relevant
```

```python

# Standardize player names by stripping any leading/trailing whitespace
player_valuations_relevant['Player'] = player_valuations_relevant['Player'].str.strip()
player_stats_2021_2022['Player'] = player_stats_2021_2022['Player'].str.strip()
player_stats_2022_2023['Player'] = player_stats_2022_2023['Player'].str.strip()

player_valuations_relevant
```

```python

# Extract relevant columns: player names and market valuations
player_valuations_relevant = player_valuations_with_names[['name','date' , 'market_value_in_eur']]
player_valuations_relevant.columns = ['Player', 'date', 'Valuation']
player_valuations_relevant
```

```python

# Standardize player names by stripping any leading/trailing whitespace
player_valuations_relevant['Player'] = player_valuations_relevant['Player'].str.strip()
player_stats_2021_2022['Player'] = player_stats_2021_2022['Player'].str.strip()
player_stats_2022_2023['Player'] = player_stats_2022_2023['Player'].str.strip()

player_valuations_relevant
```

```python
player_valuations_relevant['date'] = pd.to_datetime(player_valuations_relevant['date'])

# extract only the rows that are dated between 2021-8-1 and 2022-6-1
player_valuations_relevant_21_22 = player_valuations_relevant[(player_valuations_relevant['date'] >= '2021-8-1') & (player_valuations_relevant['date'] <= '2022-6-1')]

# extract only the rows that are dated between 2022-8-1 and 2023-6-1
player_valuations_relevant_22_23 = player_valuations_relevant[(player_valuations_relevant['date'] >= '2022-8-1') & (player_valuations_relevant['date'] <= '2023-6-1')]

```

```python
# if there are multiple valuations for a player in a season, we will keep the most recent valuation
player_valuations_relevant_21_22 = player_valuations_relevant_21_22.sort_values('date').drop_duplicates('Player', keep='last')
player_valuations_relevant_22_23 = player_valuations_relevant_22_23.sort_values('date').drop_duplicates('Player', keep='last')
player_valuations_relevant_22_23
```

```python

# Merging player valuations with player stats for 2021-2022
merged_2021_2022 = player_stats_2021_2022.merge(player_valuations_relevant_21_22, on='Player', how='left')

# Merging player valuations with player stats for 2022-2023
merged_2022_2023 = player_stats_2022_2023.merge(player_valuations_relevant_22_23, on='Player', how='left')

# combine the two dataframes to get a single dataframe and save it to a CSV file
combined = pd.concat([merged_2021_2022, merged_2022_2023])

# drop any rows that have missing values in the 'Valuation' column
combined = combined.dropna(subset=['Valuation'])


combined.to_csv('processed_data/combined.csv', index=False)
```

```python
# Get the unique values in the 'Pos' column
unique_positions = combined['Pos'].unique()

# Split the dataframe and save each subset to a separate CSV file
for position in unique_positions:
    subset_df = combined[combined['Pos'] == position]
    output_path = f'processed_data/{position}_players.csv'
    # drop all columns that have string values
    # subset_df = subset_df.drop(columns=['Player', 'Club', '  .... 
    subset_df.to_csv(output_path, index=False)

```
