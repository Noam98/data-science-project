```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Load the CSV files with appropriate encoding and delimiter
player_stats_2021_2022 = pd.read_csv('kaggle_raw_data/2021-2022 Football Player Stats.csv', encoding='latin1', delimiter=';')
player_stats_2022_2023 = pd.read_csv('kaggle_raw_data/2022-2023 Football Player Stats.csv', encoding='latin1', delimiter=';')
player_valuations = pd.read_csv('kaggle_raw_data/player_valuations.csv', encoding='latin1')
players = pd.read_csv('kaggle_raw_data/players.csv', encoding='latin1')

player_stats_2021_2022.head(), player_stats_2022_2023.head(), player_valuations.head(), players.head()
```

```python
# Merge player valuations with players to get player names and positions
player_valuations_with_names = player_valuations.merge(players[['player_id', 'name', 'sub_position']], on='player_id', how='left')

# Extract relevant columns: player names, dates, positions and market values
player_valuations_relevant = player_valuations_with_names[['name', 'date', 'sub_position', 'market_value_in_eur']]
player_valuations_relevant.columns = ['Player','Date','Position', 'Valuation']
player_valuations_relevant
```

```python
# Standardize player names by stripping any leading/trailing whitespace
player_valuations_relevant['Player'] = player_valuations_relevant['Player'].str.strip()
player_stats_2021_2022['Player'] = player_stats_2021_2022['Player'].str.strip()
player_stats_2022_2023['Player'] = player_stats_2022_2023['Player'].str.strip()

player_valuations_relevant
```

```python
player_valuations_relevant['Date'] = pd.to_datetime(player_valuations_relevant['Date'])

# extract only the rows that are dated between 2021-8-1 and 2022-6-1
player_valuations_relevant_21_22 = player_valuations_relevant[(player_valuations_relevant['Date'] >= '2021-8-1') & (player_valuations_relevant['Date'] <= '2022-6-1')]

# extract only the rows that are dated between 2022-8-1 and 2023-6-1
player_valuations_relevant_22_23 = player_valuations_relevant[(player_valuations_relevant['Date'] >= '2022-8-1') & (player_valuations_relevant['Date'] <= '2023-6-1')]

player_valuations_relevant_21_22.head(), player_valuations_relevant_22_23.head()
```

```python
# if there are multiple valuations for a player in a season, we will keep the most recent valuation
player_valuations_relevant_21_22 = player_valuations_relevant_21_22.sort_values('Date').drop_duplicates('Player', keep='last')
player_valuations_relevant_22_23 = player_valuations_relevant_22_23.sort_values('Date').drop_duplicates('Player', keep='last')

player_valuations_relevant_22_23.head(), player_valuations_relevant_21_22.head()
```

```python
# Merging player valuations with player stats 
merged_2021_2022 = player_stats_2021_2022.merge(player_valuations_relevant_21_22, on='Player', how='left')
merged_2022_2023 = player_stats_2022_2023.merge(player_valuations_relevant_22_23, on='Player', how='left')

# Find common columns between the two DataFrames
common_columns = merged_2021_2022.columns.intersection(merged_2022_2023.columns)

# Select only the common columns from both DataFrames
merged_2021_2022_common = merged_2021_2022[common_columns]
merged_2022_2023_common = merged_2022_2023[common_columns]

merged_2021_2022_common.head(), merged_2022_2023_common.head()
```

```python
# combine the two dataframes to get a single dataframe and save it to a CSV file
combined = pd.concat([merged_2021_2022_common, merged_2022_2023_common])

# drop any rows that have missing values in the 'Valuation' column
combined = combined.dropna(subset=['Valuation'])

# drop any rows that have irrelevant string values
irrelevant_columns = ['Rk', 'Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Born'] # we have Age so Born is irrelevant
combined = combined.drop(columns=irrelevant_columns)

# narrow down the positions to only the main positions
combined['Position'] = combined['Position'].str.replace('Left Winger', 'Winger', regex=False)\
                                .str.replace('Right Winger', 'Winger', regex=False)\
                                .str.replace('Central Midfield', 'Defensive Midfield', regex=False)\
                                .str.replace('Left Midfield', 'Attacking Midfield', regex=False)\
                                .str.replace('Right Midfield', 'Attacking Midfield', regex=False)\
                                .str.replace('Left-Back', 'Back', regex=False)\
                                .str.replace('Right-Back', 'Back', regex=False)\
                                .str.replace('Second Striker', 'Centre-Forward', regex=False)
                                
combined.to_csv('processed_data/combined.csv', index=False)
combined.head()
```

```python
# Get the unique values in the 'position' column
unique_positions = combined['Position'].unique()

# Split the dataframe and save each subset to a separate CSV file
for position in unique_positions:
    subset_df = combined[combined['Position'] == position]
    output_path = f'processed_data/{position}_players.csv'
    subset_df.to_csv(output_path, index=False)
    print(f'{position}_players.csv - {len(subset_df)} rows')


```
